```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
```

```python
df = pd.read_csv('../data/stock_GE.csv')
```

```python
df.head()
```

```python
len(df)
```

```python
prices_log_ratios = np.log(df["Open"].iloc[1:].values/df["Open"].iloc[:-1].values)
```

```python
selected_prices = prices_log_ratios[len(prices_log_ratios)-1000:len(prices_log_ratios)]
```

```python
len(selected_prices)
```

```python
np.save("../data/GE_observations.npy", selected_prices)
```

```python
fig, ax = plt.subplots(figsize=(40,20))
x = list(range(len(selected_prices)))
ax.plot(x, selected_prices)
#labels = df["Date"]
# You can specify a rotation for the tick labels in degrees or with keywords.
#plt.xticks(x, labels, rotation='vertical',fontsize=8)
#plt.subplots_adjust(bottom=0.15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
